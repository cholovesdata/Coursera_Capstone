```python
# import libraries
import pandas as pd
import numpy as np
```

```python
wikipedia_page_link = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"

# scrape data from html tables
# references
# https://www.marsja.se/how-to-use-pandas-read_html-to-scrape-data-from-html-tables/
# https://beenje.github.io/blog/posts/parsing-html-tables-in-python-with-pandas/
df = pd.read_html(wikipedia_page_link)[0]
df
```

```python
# The dataframe will consist of three columns: PostalCode, Borough, and Neighborhood
# Change dataframe column names
df = df.rename(columns={"Postal Code": "PostalCode"})
df
```

```python
# Only process cells that have an assigned borough. Ignore cells with a borough that is Not assigned
df = df[df['Borough'] != "Not assigned"]
df = df.reset_index(drop=True)
df
```

```python
# More than one neighborhood can exist in one postal code area. 
# For example, in the table on the Wikipedia page, you will notice that M5A is listed twice and has two neighborhoods: Harbourfront and Regent Park.
# These two rows will be combined into one row with the neighborhoods separated with a comma as shown in row 11 in the above table.

# the dataframe does not have duplicate PostalCode
df[df["PostalCode"] == "M5A"]
```

```python
# If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough.
df[df["Neighborhood"] == "Not assigned"]
# the dataframe doesn't have "Not assigned" neighborhood
```
